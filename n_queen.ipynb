```python
import numpy as np
import random as rd


class Chromosome():
      def __init__(self):
          sequence = None
          fitness = None
      def setSequence(self, val):
        self.sequence = val
      def setFitness(self, fitness):
        self.fitness = fitness


def fitness(chromosome):
    temp = chromosome.copy()
    index_start = 0
    fitness_value = 0

    # Horizontal
    len_set = len(set(chromosome))
    len_list = len(chromosome)

    fitness_value += len_list - len_set

    # Diagonal
    while len(temp) != 0:
      start = temp.pop(0)
      for index in range(index_start+1, len(chromosome)):
        invert_diagonal = index + chromosome[index] == index_start + start
        normal_diagonal = index - chromosome[index] == index_start - start

        if normal_diagonal or invert_diagonal:
          fitness_value += 1

      index_start += 1

    return fitness_value


def orderGeneration(population):
    population.sort(key=lambda x: x.fitness)

    return population


def generatePopulation():
    population = [Chromosome() for i in range(1000)]
    for i in range(1000):
      population[i].setSequence(rd.sample(range(0, 8), 8))
      population[i].setFitness(fitness(population[i].sequence))

    return population


def getParents(population, childrens_value):
    parents_arr = []
    i = 0
    while i < childrens_value:
      one, two = rd.choices(population, weights=[fitness.fitness for fitness in population], k=2)

      if one.sequence != two.sequence:
        parents_arr.append([one, two])

        i += 1

    return parents_arr


def crossover(parents):
  childrens = []

  for parent in parents:
    value = rd.randint(0, 7)
    one, two = parent

    new_one = Chromosome()
    new_one.setSequence(one.sequence[:value] + two.sequence[value:])
    new_one.setFitness(fitness(new_one.sequence))

    new_two = Chromosome()
    new_two.setSequence(two.sequence[:value] + one.sequence[value:])
    new_two.setFitness(fitness(new_two.sequence))

    childrens.append(new_one)
    childrens.append(new_two)

  return childrens


def mutation(population):
  mutation_porcent = 5

  for p in population:
    value = rd.randint(0,100)

    if value <= mutation_porcent:
      index = rd.randint(0, 7)
      p.sequence[index] = rd.randint(0, 7)
      p.setFitness(fitness(p.sequence))

  return population


if __name__ == '__main__':
    reset = True
    while reset:
      population = generatePopulation()
      population = orderGeneration(population)

      result = None
      childrens_value = len(population) // 2

      while result == None:
        parents = getParents(population, childrens_value)
        population = crossover(parents)
        population = mutation(population)
        
        for p in population:
          if p.fitness == 0:
            result = p

        childrens_value = (childrens_value // 2)
        if childrens_value <= 2: break
      
      if result != None:
        print(result.sequence)
        reset = False
```
